```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

## Categorical

### Bar Plots (Single Category)

```python
sns.countplot( x=df["room_type"])
```

```python
sns.countplot(data=df, x= "neighbourhood_group")
```

```python
sns.countplot(data = df, x ="neighbourhood_group" , hue = "room_type") #hue is used to differentiate the data
```

### Pie Chart

```python
df["room_type"].value_counts().plot(kind = "pie", autopct = "%.2f") #autopct is used to show the percentage
```

```python
df["availability_365"].value_counts().head(10).plot(kind ="bar") # top 10 availability
```
