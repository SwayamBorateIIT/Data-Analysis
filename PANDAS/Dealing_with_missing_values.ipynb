```python
data.isnull().sum()
```

## Method 1 : Removing Column

```python
df1 = data.drop("last_review",axis =1) # drop last_review column
df1.isnull().sum()
```

```python
df1.shape
```

## Method 2: Removing Rows

```python
df2 = data.dropna() # drop all rows with missing values
df2.shape
```

```python
df2.isna().sum()
```

## Method 3: Filling the missing values - Imputation

```python
df3 = data.copy()

df3.shape
```

```python
df3.head()
```

```python
df3.isna().sum()
```

```python
mean_value = df3['reviews_per_month'].mean()

df3['reviews_per_month'].fillna(mean_value, inplace=True)
# df3['reviews_per_month'] = df3['reviews_per_month'].fillna(mean_value)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

## Categorical Data (Using mode values to replace)

```python
df4=data.copy()
df4["last_review"].fillna(df4["last_review"].value_counts().index[0],inplace=True) # fill with most frequent value
df4.head()
```

```python
df4.isna().sum()
```

## Categorical Data (Using custom input to replace)

```python
df5=data.copy()
df5["last_review"].fillna("Not Reviewed", inplace = True) # fill with a string
df5.head()
```

```python
df5["host_name"].fillna("Unknown", inplace = True)
```

```python
df5.isnull().sum()
```

## Advanced Imputation

```python
df6=data.copy()
df6["reviews_per_month"].interpolate(inplace = True)
```

```python
df6.head()
```

```python
df6.isnull().sum()
```
