## sns.distplot()

#### sns.distplot(): This function from the Seaborn library is used to create a distribution plot. It visualizes the distribution of a continuous variable, combining both a histogram and a kernel density estimate (KDE). The histogram shows the frequency distribution, while the KDE represents the probability density function.


```python
sns.distplot(df2["age"])
```

```python
sns.distplot(df2["fare"])
```

## Boxplot
#### A boxplot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset. It displays the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values of a dataset, helping to visualize the spread and skewness of the data.

```python
sns.boxplot(df2["age"])
```

## Violin Plot

#### A boxplot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset. It displays the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values of a dataset, helping to visualize the spread and skewness of the data.

```python
sns.violinplot(df2["age"])
```
