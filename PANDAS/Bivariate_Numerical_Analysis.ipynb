

# Jointplot in Seaborn

A **`jointplot`** in Seaborn is a function used to create a **multi-dimensional plot** that combines a scatter plot (or other types of bivariate plots) with the marginal distributions of the individual variables. It provides a detailed view of the relationship between two variables.

## Key Features

### Central Plot:
- Displays the relationship between two variables (e.g., a scatter plot, regression plot, or KDE plot).

### Marginal Plots:
- Adds histograms, KDE plots, or other univariate visualizations along the x-axis and y-axis to show the distribution of each variable.

### Customizable `kind`:
The type of plot used for the central visualization can be specified using the `kind` parameter:
- `scatter`: Scatter plot (default).
- `reg`: Scatter plot with a regression line.
- `kde`: Kernel density estimate for the joint distribution.
- `hex`: Hexbin plot for large datasets.



```python
sns.jointplot(x= "total_bill", y= "tip",hue = "smoker" ,data = df2) 
# jointplot is a scatterplot with proability density function
```

```python
sns.jointplot(x= "total_bill", y= "tip",data = df2, kind ="reg")
# kind = "reg" will add a regression line to the scatterplot
```



## Correlation Matrix

A **correlation matrix** is a table that shows the correlation coefficients between multiple variables in a dataset. It is a way to summarize the strength and direction of linear relationships between pairs of variables.

## Key Features

### Correlation Coefficients:
- Values range from **-1 to 1**:
  - **1**: Perfect positive correlation.
  - **0**: No correlation.
  - **-1**: Perfect negative correlation.
- Positive values indicate that as one variable increases, the other tends to increase.
- Negative values indicate that as one variable increases, the other tends to decrease.

### Symmetric Matrix:
- The matrix is symmetric because the correlation between `X` and `Y` is the same as between `Y` and `X`.

### Diagonal Values:
- The diagonal of the matrix always contains **1** because each variable is perfectly correlated with itself.


```python
df2.corr(numeric_only=True) # correlation matrix
```

## LinePlot 

### Seaborn `sns.lineplot`

In Seaborn, `sns.lineplot` is a function used to create **line plots**, which are useful for visualizing data trends over time or showing relationships between two variables.

### Key Features of `sns.lineplot`:

### X and Y Axes:
- Displays one variable (often continuous) on the x-axis and another on the y-axis.
- The plot connects data points with a line to highlight trends.

### Handles Categories:
- Can include a categorical variable for `hue` to show multiple lines for different categories.

### Automatic Confidence Interval (CI):
- By default, a shaded area around the line shows a **confidence interval**, which can be disabled with `ci=None`.

### Customizable:
- Supports various styling options for markers, line types, and colors.


```python
sns.lineplot(x = "Date", y= "Price", data = df)
```

```python
sns.lineplot(x = "Date", y= "Price", data = df.head(10))
plt.xticks(rotation = 90)
plt.show()
```
