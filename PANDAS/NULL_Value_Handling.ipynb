```python
import pandas as pd
import numpy as np


from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('data.csv')
df
```

## Working on the NULL value with Numerical data

## Removing the rows with NULL values 

```python
df.dropna()
```

## Replacing the NULL values with the mean of that column uisng Imputers

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')

# Creates a SimpleImputer object to replace missing values (np.nan) in the data.
# The imputer is initialized with the strategy='mean' parameter, meaning missing values will be replaced by the mean of the column where they occur.
```

```python
imputer.fit(df[['Salary', 'Age']])  # The fit() method calculates the mean for each of the columns ('Salary' and 'Age') in the DataFrame df.
df[['Salary', 'Age']] = imputer.transform(df[['Salary', 'Age']]) # The transform() method replaces any missing values (np.nan) in the 'Salary' and 'Age' columns with the means calculated during the fit() step.
df
```

## Working on the NULL values with categorical data

## Replacing the NULL countries with the most frequent country

```python
imputer=SimpleImputer(missing_values=np.nan, strategy='most_frequent') # Creates a new SimpleImputer object with the strategy='most_frequent' parameter, meaning missing values will be replaced by the most frequent value in the column where they occur.

imputer.fit(df[['Country', 'Purchased']]) # The fit() method calculates the most frequent value for each of the columns ('Country' and 'Purchased') in the DataFrame df.

df[['Country', 'Purchased']] = imputer.transform(df[['Country', 'Purchased']]) # The transform() method replaces any missing values (np.nan) in the 'Country' and 'Purchased' columns with the most frequent values calculated during the fit() step.
df
```
