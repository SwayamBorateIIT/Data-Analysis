```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()

```

## Barplot

### Seaborn `sns.barplot`

In Seaborn, `sns.barplot` is a function used to create **bar plots**, which represent data in rectangular bars to show comparisons between different categories or groups.

### Key Features of `sns.barplot`:

### Categorical Data on X-Axis:
- Plots categories on the x-axis and a numerical value on the y-axis.

### Height of Bars:
- Each bar represents the **mean (average)** of the numerical values for that category by default.

### Error Bars:
- Displays **confidence intervals** or **standard deviation** as error bars, providing a visual indication of variability.

### Customizable:
- Allows customization of colors, error bar type, orientation, and more.


```python
sns.barplot(x ="class", y= "age", data = df)
```

```python
sns.barplot(x ="class", y= "age",hue = "sex", data = df) 
# hue is used to differentiate the data
```

```python
sns.barplot(x ="class", y= "age",hue = "sex",estimator="sum", data = df) 
# estimator is used to change the estimator function
```

```python
sns.boxplot(x = "sex", y= "age", data = df)
# boxplot is used to show the distribution of the data
```

```python
sns.boxplot(x = "sex", y= "age", hue = "class", data = df)
# hue is used to differentiate the data
```

```python
sns.distplot(df[df["survived"]==0]["age"], hist = False)
sns.distplot(df[df["survived"]==1]["age"], hist = False)
```

```python

sns.violinplot(x = "sex", y = "fare", data = df)
```
